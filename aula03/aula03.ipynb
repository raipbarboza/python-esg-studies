```python
# Combinando filtros — & é o AND, | é o OR
df[(df["score_esg"] > 70) & (df["setor"] == "Consumo")]
```

```python
# Coluna de classificação — aplicando a lógica da Aula 1
def classificar(score):
    if score >= 80:
        return "Líder"
    elif score >= 60:
        return "Avançado"
    else:
        return "Básico"

df["classificacao"] = df["score_esg"].apply(classificar)
df
```

```python
# Total de emissões por setor
df.groupby("setor")["emissoes"].sum()
```

```python
# Score médio por setor
df.groupby("setor")["score_esg"].mean()
```

```python
# Várias agregações de uma vez
df.groupby("setor").agg(
    total_emissoes=("emissoes", "sum"),
    score_medio=("score_esg", "mean"),
    qtd_empresas=("nome", "count")
)
```
