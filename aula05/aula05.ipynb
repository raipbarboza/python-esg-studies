```python
df_ise = pd.read_csv(
    r"C:\carbontrack-br\data\raw\ise_carteira.csv",
    sep=";",
    encoding="latin1",
    skiprows=1,
    skipfooter=2,
    engine="python",
    names=["codigo", "acao", "tipo", "qtde_teorica", "participacao_pct", "vazio"],
    header=0
)

# Remover coluna vazia
df_ise = df_ise.drop(columns=["vazio"])

print(df_ise.shape)
print(df_ise.columns.tolist())
df_ise.head(10)
```

```python
# Salvar ISE limpo
df_ise.to_csv(
    r"C:\carbontrack-br\data\processed\ise_limpo.csv",
    index=False,
    sep=";",
    encoding="utf-8-sig"
)
print("ISE salvo!")

# Carregar CVM que você já limpou
df_cvm = pd.read_csv(
    r"C:\carbontrack-br\data\processed\empresas_mg_esg.csv",
    sep=";",
    encoding="utf-8-sig"
)

# Cruzamento — empresas que estão nos dois datasets
# ISE usa código de ação (PETR4), CVM usa CNPJ
# Vamos cruzar pelo nome da empresa
df_ise["acao_limpa"] = df_ise["acao"].str.strip().str.upper()
df_cvm["nome_limpo"] = df_cvm["DENOM_SOCIAL"].str.strip().str.upper()

# Ver as empresas ISE de MG
empresas_ise = df_ise["acao_limpa"].tolist()
print(f"\nEmpresas no ISE: {len(empresas_ise)}")
print(empresas_ise)
```

```python
# Carregar o cadastro CVM completo (não só MG)
df_cvm_completo = pd.read_csv(
    r"C:\carbontrack-br\data\raw\cad_cia_aberta.csv",
    sep=";",
    encoding="latin1"
)

# Filtrar só ativas
df_cvm_ativas = df_cvm_completo[df_cvm_completo["SIT"] == "ATIVO"].copy()
df_cvm_ativas["nome_limpo"] = df_cvm_ativas["DENOM_SOCIAL"].str.strip().str.upper()

# Cruzamento por nome — buscar cada empresa ISE na CVM
matches = []

for _, empresa in df_ise.iterrows():
    nome_ise = empresa["acao_limpa"]
    
    # Buscar na CVM empresas que contenham o nome ISE
    encontradas = df_cvm_ativas[df_cvm_ativas["nome_limpo"].str.contains(nome_ise, na=False)]
    
    if len(encontradas) > 0:
        for _, cvm in encontradas.iterrows():
            matches.append({
                "codigo_ise": empresa["codigo"],
                "nome_ise": nome_ise,
                "nome_cvm": cvm["DENOM_SOCIAL"],
                "cnpj": cvm["CNPJ_CIA"],
                "setor": cvm["SETOR_ATIV"],
                "uf": cvm["UF"],
                "participacao_pct": empresa["participacao_pct"]
            })

df_cruzado = pd.DataFrame(matches)
print(f"Empresas ISE encontradas na CVM: {df_cruzado['codigo_ise'].nunique()}")
print(f"Total de registros: {len(df_cruzado)}")
df_cruzado.head(15)
```

```python
# Ver as empresas de MG no cruzamento
df_ise_mg = df_cruzado[df_cruzado["uf"] == "MG"].copy()
print(f"Empresas ISE em MG: {len(df_ise_mg)}")
print(df_ise_mg[["codigo_ise", "nome_ise", "nome_cvm", "setor"]].to_string())
```

```python
# Salvar cruzamento completo
df_cruzado.to_csv(
    r"C:\carbontrack-br\data\processed\ise_x_cvm.csv",
    index=False,
    sep=";",
    encoding="utf-8-sig"
)

# Salvar só MG
df_ise_mg.to_csv(
    r"C:\carbontrack-br\data\processed\ise_mg.csv",
    index=False,
    sep=";",
    encoding="utf-8-sig"
)

print("Arquivos salvos!")
print(f"\nResumo do CarbonTrack BR até agora:")
print(f"Empresas CVM ativas:         {len(df_cvm_ativas)}")
print(f"Empresas ISE (certificadas): {df_cruzado['codigo_ise'].nunique()}")
print(f"Empresas ISE em MG:          {len(df_ise_mg)}")
```
