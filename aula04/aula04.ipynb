```python
colunas_uteis = [
    "CNPJ_CIA",
    "DENOM_SOCIAL",
    "DENOM_COMERC",
    "DT_REG",
    "SIT",
    "SETOR_ATIV",
    "CONTROLE_ACIONARIO",
    "UF",
    "PAIS",
    "EMAIL",
    "AUDITOR"
]

df_limpo = df[colunas_uteis]
df_limpo.head()
```

```python
df_limpo["SIT"].value_counts()
```

```python
# Filtrar só as ativas
df_ativas = df_limpo[df_limpo["SIT"] == "ATIVO"]
print(f"Total de empresas ativas: {len(df_ativas)}")
df_ativas.head()
```

```python
# Preencher nome comercial vazio com o nome social
df_ativas["DENOM_COMERC"] = df_ativas["DENOM_COMERC"].fillna(df_ativas["DENOM_SOCIAL"])
```

```python
df_ativas["DENOM_COMERC"] = df_ativas["DENOM_COMERC"].fillna(df_ativas["DENOM_SOCIAL"])
df_ativas["SETOR_ATIV"] = df_ativas["SETOR_ATIV"].fillna("Não informado")
```
