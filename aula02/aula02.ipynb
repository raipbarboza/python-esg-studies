```python
# 2 Calcule e imprima o total de emissões somando todas as empresas. 
# Dica: crie uma variável total = 0 antes do for e vá somando.

empresas = [
    {"nome": "Petrobras", "setor": "Energia",    "emissoes": 45_800_000, "score_esg": 58},
    {"nome": "Vale",      "setor": "Mineração",  "emissoes": 16_300_000, "score_esg": 72},
    {"nome": "Natura",    "setor": "Consumo",    "emissoes": 890_000,    "score_esg": 91},
    {"nome": "Itaú",      "setor": "Financeiro", "emissoes": 1_200_000,  "score_esg": 85},
    {"nome": "Ambev",     "setor": "Consumo",    "emissoes": 3_400_000,  "score_esg": 63},
]


total = 0

for empresa in empresas:
    emissao_atual = empresa["emissoes"]
    total = total + emissao_atual
    
print(f"Total de emissões de todas as empresas: {total:,} toneladas de CO2")
```

```python
# 3 Encontre e imprima o nome da empresa com o maior score ESG. 
# Dica: mesma lógica — crie melhor = empresas[0] antes do for 
# e compare dentro do loop.

empresas = [
    {"nome": "Petrobras", "setor": "Energia",    "emissoes": 45_800_000, "score_esg": 58},
    {"nome": "Vale",      "setor": "Mineração",  "emissoes": 16_300_000, "score_esg": 72},
    {"nome": "Natura",    "setor": "Consumo",    "emissoes": 890_000,    "score_esg": 91},
    {"nome": "Itaú",      "setor": "Financeiro", "emissoes": 1_200_000,  "score_esg": 85},
    {"nome": "Ambev",     "setor": "Consumo",    "emissoes": 3_400_000,  "score_esg": 63},
]


melhor = empresas[0]

for empresa in empresas:
    if empresa["score_esg"] > melhor["score_esg"]:
        melhor = empresa
        
print(f"Empresa com maior score ESG: {melhor['nome']}")
print(f"Score ESG: {melhor['score_esg']}")
```

```python
# 4. Adicione um novo campo "classificacao" em cada empresa 
# com os mesmos critérios da Aula 1 (Líder, Avançado, Básico) 
# e imprima nome + classificação de todas.

empresas = [
    {"nome": "Petrobras", "setor": "Energia",    "emissoes": 45_800_000, "score_esg": 58},
    {"nome": "Vale",      "setor": "Mineração",  "emissoes": 16_300_000, "score_esg": 72},
    {"nome": "Natura",    "setor": "Consumo",    "emissoes": 890_000,    "score_esg": 91},
    {"nome": "Itaú",      "setor": "Financeiro", "emissoes": 1_200_000,  "score_esg": 85},
    {"nome": "Ambev",     "setor": "Consumo",    "emissoes": 3_400_000,  "score_esg": 63},
]
for empresa in empresas:
    score = empresa["score_esg"]
    
    if score >= 80:
        empresa["classificacao"] = "Líder"
    elif score >= 60:
        empresa["classificacao"] = "Avançado"
    else:
        empresa["classificacao"] = "Básico"
    
    print(f"{empresa['nome']}: {empresa['classificacao']}")


```
