```python
# Dados da empresa
nome_empresa = "  vale s.a.  "
emissoes_scope1 = 12_500_000
emissoes_scope2 = 3_800_000
meta_reducao_pct = 30
ano_base = 2019
emissoes_ano_base = 18_000_000
soma = emissoes_scope1 + emissoes_scope2
meta_absoluta = emissoes_ano_base * (1 - meta_reducao_pct / 100)

# 1. Corrija o nome da empresa: sem espaços nas bordas 
# e com a primeira letra de cada palavra em maiúsculo. 
# Dica: procure o método .title().
# 2. Calcule o total de emissões (Scope 1 + Scope 2).
# 3. Calcule qual seria a meta absoluta de emissões para 2030 
# (ano base menos a % de redução).
# 4. Compare o total atual com a meta: a empresa já atingiu a meta? 
# Use um if para imprimir 
# "Meta atingida ✓" ou "Meta não atingida — faltam X tCO2e".

# Resultados (1,2,3)
print(nome_empresa.strip().title())
print(f"Total de Emissões: {soma:,} Toneladas")
print(f"Meta Absoluta 2030: {meta_absoluta:,} Toneladas")

# Verificação da meta (4)
if soma <= meta_absoluta:
    print("Meta atingida ✓")
else:
    diferenca = soma - meta_absoluta
    print(f"Meta não atingida — faltam {diferenca:,} tCO2e")
```
